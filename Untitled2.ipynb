```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1 = cv2.imread('2222.png',cv2.IMREAD_GRAYSCALE)          # queryImage
img2 = cv2.imread('1111.png',cv2.IMREAD_GRAYSCALE)          # trainImage
```

```python
scale=51
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
width=int(img1.shape[1]*scale/100)
height=int(img1.shape[0]*scale/100)
dim=(width,height)
img1=cv2.resize(img1,dim,interpolation=cv2.INTER_AREA)
plt.imshow(img1)
print('Image shape: ', np.shape(img1))
```

```python
scale=51
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
width=int(img2.shape[1]*scale/100)
height=int(img2.shape[0]*scale/100)
dim=(width,height)
img2=cv2.resize(img2,dim,interpolation=cv2.INTER_AREA)
plt.imshow(img2)
print('Image shape: ', np.shape(img2))
```

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1,des2)
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:20],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
```
