## Problem 1
___
Load and show the image and print shape of the image.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1=cv2.imread('pic_o.jpg')
print(img1.shape)
plt.imshow(img1)
```

```python
scale=7.03
width=int(img1.shape[1]*scale/100)
height=int(img1.shape[0]*scale/100)
dim=(width,height)
img=cv2.resize(img1,dim,interpolation = cv2.INTER_AREA)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
print('Shape of image: ', np.shape(img))
```
