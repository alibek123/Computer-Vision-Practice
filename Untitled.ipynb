```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import cv2

img=cv2.imread('moon.jpg',1)

cropped_image=img[180:300,190:310]

height, width = (cropped_image.shape[0],cropped_image.shape[1])

x_offset = int((width - cropped_image.shape[1])/2)
y_offset = int((height - cropped_image.shape[0])/2)

img[y_offset:y_offset+cropped_image.shape[0], x_offset:x_offset+cropped_image.shape[1]]=cropped_image
plt.imshow(img)
cv2.imwrite('C:/Users/zhuma/Desktop/CV prac/2_moon.jpg',img)
```
