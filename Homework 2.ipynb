## Problem 2

a) Blur your image with kernel size 5x5, 10x10 and 2x5. Show the result.

b) Blur half part of your image (kernal size 15x15). Show the result.

c) Take any piece of the image.Blur this piece and select this piece in the image in the green frame. Show the result.

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread('pic_o.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
scale=7.03
width=int(img.shape[1]*scale/100)
height=int(img.shape[0]*scale/100)
dim=(width,height)
img1=cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
plt.imshow(img1)
print('Image shape: ', np.shape(img1))
```

```python
k_w1=5
k_h1=5

k_w2=10
k_h2=10

k_w3=2
k_h3=5

K1=np.ones((k_w1,k_h1),np.float32)/(k_w1*k_h1)
img_bl1=cv2.filter2D(img1,-1,K1)

K2=np.ones((k_w2,k_h2),np.float32)/(k_w2*k_h2)
img_bl2=cv2.filter2D(img1,-1,K2)

K3=np.ones((k_w3,k_h3),np.float32)/(k_w3*k_h3)
img_bl3=cv2.filter2D(img1,-1,K3)

plt.imshow(img_bl1)
plt.imshow(img_bl2)
plt.imshow(img_bl3)
```
